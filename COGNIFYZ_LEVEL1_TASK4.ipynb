## Check Unique Values in Has Online delivery

```python
print("Unique Values:", data['Has Online delivery'].unique())
```

## Calculate Percentage of Restaurants That Offer Online Delivery

```python
# Total number of restaurants
total = len(data)

# Count restaurants that offer online delivery
online_delivery_count = data[data['Has Online delivery'] == 'Yes'].shape[0]

# Calculate percentage
online_delivery_percentage = round((online_delivery_count / total * 100), 2)

print(f"Percentage of restaurants that offer online delivery: {online_delivery_percentage}%")
```

## Compare Average Ratings

```python
# Group by 'Has Online delivery' and calculate average rating
avg_rating_comparison = data.groupby('Has Online delivery')['Aggregate rating'].mean().round(2)

print("Average ratings by online delivery status:")
print(avg_rating_comparison)
```

## Bar Chart Comparison

```python
# Plot a bar chart to compare ratings
avg_rating_comparison.plot(kind='bar', color=['tomato', 'lightgreen'])

plt.title('Average Ratings: With vs Without Online Delivery')
plt.ylabel('Average Rating')
plt.xlabel('Online Delivery')
plt.xticks(rotation=0)
plt.ylim(0, 5)
plt.show()
```
